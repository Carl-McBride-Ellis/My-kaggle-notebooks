## How about this:
$$\Huge E=mc^2$$
### or this (the step size in [gradient descent](https://en.wikipedia.org/wiki/Gradient_descent)): 
$$\large \gamma_{n} = \frac{ \left | \left (\mathbf x_{n} - \mathbf x_{n-1} \right )^T \left [\nabla F (\mathbf x_{n}) - \nabla F (\mathbf x_{n-1}) \right ] \right |}{\left \|\nabla F(\mathbf{x}_{n}) - \nabla F(\mathbf{x}_{n-1}) \right \|^2}$$

[kaggle notebooks](https://jupyter-notebook.readthedocs.io/en/stable/examples/Notebook/Working%20With%20Markdown%20Cells.html#LaTeX-equations) markdown cells use [MathJax 'Beautiful and accessible math in all browsers'](https://www.mathjax.org/), which in turn uses [\\( \LaTeX \\) equation markup](https://en.wikipedia.org/wiki/Help:Displaying_a_formula). Wikipedia (Mediawiki) also uses \\( \LaTeX \\) in between `<math>` and `</math>`, so any equation you see on Wikipedia can be cut and pasted directly into your kaggle notebook (which I did with the equation above). You can use \\( \LaTeX \\) when [writing mathematical expressions in matplotlib](https://matplotlib.org/3.1.1/tutorials/text/mathtext.html), for example in the titles of your plots.
It is also worth noting that the vast majority of the papers that are published on [arXiv.org](https://arxiv.org/) are written in \\( \LaTeX \\) too.
### Useage:
Inline equations are surrounded by dollars symbols, and equations having their own line are surrounded by double dollars *i.e.*:`$$.`

Here are a few examples:

* Using a [Greek letter](https://en.wikipedia.org/wiki/Wikipedia:LaTeX_symbols) and an exponent: `$$ A = \pi r^2 $$`, which gives:

$$\large A = \pi r^2$$

* here is a fraction: `$$ a  = \frac {b} { \sqrt{2c} } $$`, which gives:

$$\large a  = \frac {b} {\sqrt{2c}}$$

* and here is a summation, using both subscripts and superscripts: `$$ A = \sum_{1}^{n+1} \sigma (a_i +  \delta b_i) $$`, which gives:

$$\large A = \sum_{1}^{n+1} \sigma (a_i +  \delta b_i) $$

* an integral, `$$ \ln a = \int_1^a \frac{1}{x} \, dx $$` where `\,` acts as a spacer:

$$\large \ln a = \int_1^a \frac{1}{x}\,dx $$

* hats, bars and vectors `$$ \hat a  \bar b  \vec c $$`:

$$\large \hat a  \bar b  \vec c $$

* a familiar favorite on kaggle: the the root of the [mean squared logarithmic error regression loss](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_squared_log_error.html) `$${\mathrm {RMSLE}}\,(y, \hat y) = \sqrt{\frac{1}{n_{\mathrm{samples}}}\sum_{i=0}^{n_{\mathrm{samples}}-1} \left(\ln(1+y_i) - \ln (1+ \hat y_i)\right)^2} $$`

$$ {\mathrm {RMSLE}}\,(y, \hat y) = \sqrt{ \frac{1}{n_{   \mathrm{samples}    }}  \sum_{i=0}^{n_{    \mathrm{samples} }-1} \left( \ln (1+y_i) - \ln (1+ \hat y_i) \right)^2 }  $$

* matrices and derivatives (here the [Jacobian matrix](https://en.wikipedia.org/wiki/Jacobian_matrix_and_determinant) for the spherical-Cartesian transformation):

$$\mathbf J_{\mathbf F}(r, \theta, \varphi) = \begin{bmatrix}
  \frac{\partial x}{\partial r} & \frac{\partial x}{\partial \theta} & \frac{\partial x}{\partial \varphi} \\
  \frac{\partial y}{\partial r} & \frac{\partial y}{\partial \theta} & \frac{\partial y}{\partial \varphi} \\
  \frac{\partial z}{\partial r} & \frac{\partial z}{\partial \theta} & \frac{\partial z}{\partial \varphi}\end{bmatrix}
= \begin{bmatrix}
      \sin \varphi \cos \theta & -r \sin \varphi \sin \theta &  r \cos \varphi \cos \theta \\
      \sin \varphi \sin \theta & r \sin \varphi \cos \theta & r \cos \varphi \sin \theta \\
      \cos \varphi & 0 & -r \sin \varphi 
      \end{bmatrix} $$
      
<br>
which is given by: 

`\mathbf J_{\mathbf F}(r, \theta, \varphi) = 
  \begin{bmatrix}
     \frac{\partial x}{\partial r} & \frac{\partial x}{\partial \theta} & \frac{\partial x}{\partial \varphi} \\
     \frac{\partial y}{\partial r} & \frac{\partial y}{\partial \theta} & \frac{\partial y}{\partial \varphi} \\
     \frac{\partial z}{\partial r} & \frac{\partial z}{\partial \theta} & \frac{\partial z}{\partial \varphi}   
  \end{bmatrix}
= \begin{bmatrix}
     \sin \varphi \cos \theta & -r \sin \varphi \sin \theta &  r \cos \varphi \cos \theta \\
     \sin \varphi \sin \theta & r \sin \varphi \cos \theta & r \cos \varphi \sin \theta \\
     \cos \varphi & 0 & -r \sin \varphi 
   \end{bmatrix}`

<br>
where `\mathbf` means **math bold font**, the `&` is the separator between elements, and `\\` is a line break.


For more see, for example, this page: [Wikipedia:LaTeX symbols](https://en.wikipedia.org/wiki/Wikipedia:LaTeX_symbols).
